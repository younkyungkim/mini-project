## 시간대별 총생활인구수 평균

```python
# 아래에 실습코드를 작성하고 결과를 확인합니다.


time=df_total.groupby(['시간대구분'])['총생활인구수'].mean().sort_values()
import matplotlib.pyplot as plt

# Assuming you have already computed the 'time' Series as you mentioned in your question
# If not, you should run the following line first:
# time = df_total.groupby(['시간대구분'])['총생활인구수'].mean()

# Create the bar plot
plt.figure(figsize=(10, 6))  # Adjust the figure size as needed
time.plot(kind='bar',color='skyblue')

# Add labels and title
plt.xlabel('시간대구분')
plt.ylabel('평균 총생활인구수')
plt.title('평균 총생활인구수 by 시간대구분')

# Show the plot
plt.show()
```

**역시나 출근 시간 전과 퇴근 시간 후에는 총생활인구수가 적음**

**6-18시에 생활인구수가 많음**

### 성별 총 생활 인구수 구하기

```python
# 아래에 실습코드를 작성하고 결과를 확인합니다.
man_columns = df_total[df_total.columns[df_total.columns.str.contains('남자')]]

man=man_columns.sum(axis=1)

woman_columns =  df_total[df_total.columns[df_total.columns.str.contains('여자')]]
woman=woman_columns.sum(axis=1)

df_total['여자총생활인구수'] = man
df_total['남자총생활인구수'] = woman
```

```python
# Reorder the columns
column_order = ['기준일ID', '시간대구분', '총생활인구수', '여자총생활인구수', '남자총생활인구수'] + \
               [col for col in df_total.columns if col not in ['기준일ID', '시간대구분', '총생활인구수', '여자총생활인구수', '남자총생활인구수']]

# Create a DataFrame with the reordered columns
df_total = df_total[column_order]

# Print the updated DataFrame
display(df_total.head())
```

```python
sex_mean=df_total.groupby('시간대구분')[['남자총생활인구수','여자총생활인구수']].mean().reset_index()
sex_mean.head(5)
```

```python
plt.figure(figsize=(10, 6))  # Adjust the figure size as needed
sns.barplot(x='시간대구분', y='남자총생활인구수', data=sex_mean, color='blue', label='남자', alpha=0.7)
sns.barplot(x='시간대구분', y='여자총생활인구수', data=sex_mean, color='pink', label='여자', alpha=0.7)

# Add labels and legend
plt.xlabel('시간대구분')
plt.ylabel('평균 총생활인구수')
plt.title('남자 vs. 여자 평균 총생활인구수')
plt.legend()

# Show the plot
plt.show()
```

**5년간 시간대 구분별 여성과 남성의 평균 생활인구수를 시각화하였을 때 여성이 남성보다 많은 것을 알 수 있음**

```python
year_count=[df17['총생활인구수'].mean(),df18['총생활인구수'].mean(),df19['총생활인구수'].mean(),df20['총생활인구수'].mean(),df21['총생활인구수'].mean()]

# Years corresponding to the data
years = [2017, 2018, 2019, 2020, 2021]

# Create a bar plot
plt.figure(figsize=(10, 6))  # Adjust the figure size as needed
plt.bar(years, year_count, color='skyblue')

# Add labels and title
plt.xlabel('연도')
plt.ylabel('평균 총생활인구수')
plt.title('연도별 평균 총생활인구수')

# Show the plot
plt.show()
```

**해가 거듭할수록 총생활 인구수가 점점 줄어들고 있다 -> 코로나 시기와 맞물림**

### 연령대별 평균생활인구수 

```python
man_age=man_columns.mean()
woman_age=woman_columns.mean()
```

```python
data=(man_age.values+woman_age.values)/2
age = ["0세부터9세", "10세부터14세", "15세부터19세", "20세부터24세", "25세부터29세", 
         "30세부터34세", "35세부터39세", "40세부터44세", "45세부터49세", "50세부터54세", 
         "55세부터59세", "60세부터64세", "65세부터69세", "70세 이상"]

# 데이터프레임 생성
df = pd.DataFrame(data, index=age, columns=["평균생활인구수"])

# 데이터프레임 출력
df=df.reset_index()
df = df.rename(columns={"index": "age"})
df.head(5)
```

```python
# 선 그래프 그리기
colors = ['skyblue']

plt.figure(figsize=(12, 6))
plt.plot(df['age'], df['평균생활인구수'], marker='o', linestyle='-', color=colors[0], linewidth=2, markersize=8)
plt.xlabel('연령대', fontsize=14)
plt.ylabel('평균생활인구수', fontsize=14)
plt.title('연령대별 평균생활인구수', fontsize=16)
plt.xticks(rotation=45, fontsize=12)
plt.yticks(fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)

# 그래프 주석 추가
for i, value in enumerate(df['평균생활인구수']):
    plt.annotate(f'{value:.2f}', (i, value), textcoords="offset points", xytext=(0, 10), ha='center', fontsize=10)


# 그래프 표시
plt.tight_layout()
plt.show()
```

**0세부터 9세, 70세 이상의 인구 수가 눈에 뛰게 많으며, 경제적으로 활동하는 20세-50세가 비교적 많다**
